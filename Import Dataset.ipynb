```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
from sklearn.datasets import load_iris
iris = load_iris()

# Create a DataFrame
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# Filter for setosa and versicolor
setosa = df[df['species'] == 'setosa']
versicolor = df[df['species'] == 'versicolor']

# Statistical summary
print("Setosa Summary:")
print(setosa.describe())
print("\nVersicolor Summary:")
print(versicolor.describe())

# Visualize the comparison
sns.pairplot(df[df['species'].isin(['setosa', 'versicolor'])], hue='species', diag_kind='kde')
plt.suptitle("Setosa vs Versicolor Comparison", y=1.02)
plt.show()
```
